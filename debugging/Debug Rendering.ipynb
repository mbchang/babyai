```python
import babyai
import gym
import gym_minigrid
import matplotlib
matplotlib.use('Agg')
from matplotlib import pyplot as plt
%matplotlib inline 
```

```python
def debug_rendering(env_name):
    env = gym.make(env_name)
    state = env.reset()
    image = state['image']
    frame = env.render(mode='rgb_array')
    print(frame.shape)
    plt.imshow(frame)
```

```python
debug_rendering('BabyAI-OpenRedDoor-v0')
```
